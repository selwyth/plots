```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np

plt.xkcd()

fig = plt.figure()
data = np.ones(100)
data[70:] -= np.linspace(0,1,30)

plt.annotate(
    'nz beats sa!!',
    xy=(70, 1), arrowprops=dict(arrowstyle='->'), xytext=(115, 1.5))

plt.plot(data, label = 'jing')

david = np.array(np.ones(100))* 0.5 
david[70:] = np.linspace(0.5,2.5,30)

plt.plot(david, label = 'dfeng')

ashley = np.array(np.ones(100)) * 0.25
ashley[40:70] = np.linspace(0.25,1.5,30)
ashley[70:] = np.linspace(1.5,0,30)

plt.plot(ashley, label = 'ashley')

raghav = np.array(np.ones(100)) * 0.1

plt.plot(raghav, label = 'raghav')

plt.xlabel('time')
plt.ylabel('chance of winning world cup pool')
plt.xticks([])
plt.yticks([])
plt.ylim([0,2.5])
plt.legend(loc = 'best')
```
